##Implement a Breadth-first Search Algorithm

Create a five level 'connections' graph.
- The first level should have five people and each person in the first level should have five connections (they can be exclusive to one person or shared with other members on the same level, but all five should be unique, no duplicates)
- Each person in the second level should have five connections (they can be exclusive to one person or shared with other members on the same level, but all five should be unique, no duplicates)
- Continue the same for third, fourth, and fifth levels
- There should be no reverse connections (i.e. first level can connect to second level but second cannot have any reverse connection with the first, and so on for 3rd, 4th, and 5th levels)

```python
import names
import random
import numpy as np
import pandas as pd

from collections import deque
import time
from tabulate import tabulate
import matplotlib.pyplot as plt
```

```python
#breath-first search algorithm
def search(name, level_list, element_position):
    search_queue = deque()
    search_queue += graph[name]
    # This is how you keep track of which people you've searched before.
    searched = set()
    while search_queue:
        person = search_queue.popleft()
        # Only search this person if you haven't already searched them.
        if person not in searched:
            if person == level_list[element_position]:
                print(person + " is who you're looking for!")
                return True
            else:
                search_queue += graph[person]
                # Marks this person as searched
                searched.add(person)
    return False
```

```python
def search_time(list, position):
  """ Search Time Algorithm.

  Time the search for values in the list.
  
  Parameters
  ----------
  list: list to search.
  position: position of element in the list.
  
  """
  start_time = time.perf_counter()
  search("you", list, position)
  end_time = time.perf_counter()
  total_time = end_time - start_time

  return list[position], total_time
```

```python
name_list = []
#generate random names
for i in range(4500):
  rand_name = names.get_full_name()
  name_list.append(rand_name)
```

```python
print(f"old list length: {len(name_list)}")
name_list = list(set(name_list)) #only unique strings
print(f"new list length: {len(name_list)}")
```

```python
graph = {}

#first level
graph['you'] = random.sample(name_list, 5) #select 5 random names from name_list

for name in graph["you"]:
  name_list.remove(name) #remove name from list to avoid repeats
print(f"graph['you'] = {graph['you']}") 

#choose 3 random names from first level
first_level_search = random.sample(graph["you"], 3) 
```

```python
#second level
second_level = []

for person in graph['you']:
  graph[person] = random.sample(name_list, 5)
  for name in graph[person]:
    second_level.append(name)
    name_list.remove(name)
  print(f"graph[{person}] = {graph[person]}")

#choose 3 random names from second level
second_level_search = random.sample(second_level, 3) 
```

```python
#third level
third_level = []

for i in graph['you']:
  for person in graph[i]:
    graph[person] = random.sample(name_list, 5)
    for name in graph[person]:
      third_level.append(name)
      name_list.remove(name)
    print(f"graph[{person}] = {graph[person]}")  

#choose 3 random names from third level
third_level_search = random.sample(third_level, 3) 
```

```python
#fourth level
fourth_level = []

for i in graph['you']:
  for a in graph[i]:
    for person in graph[a]:
      graph[person] = random.sample(name_list, 5)
      for name in graph[person]:
        fourth_level.append(name)
        name_list.remove(name)
      print(f"graph[{person}] = {graph[person]}")  

#choose 3 random names from fourth level
fourth_level_search = random.sample(fourth_level, 3) 
```

```python
#fifth level
fifth_level = []

for i in graph['you']:
  for a in graph[i]:
    for b in graph[a]:
      for person in graph[b]:
        graph[person] = random.sample(name_list, 5)
        for name in graph[person]:
          fifth_level.append(name)
          name_list.remove(name)
        print(f"graph[{person}] = {graph[person]}")  

#choose 3 random names from fifth level
fifth_level_search = random.sample(fifth_level, 3) 
```

```python
#terminal nodes
for i in graph['you']:
  for a in graph[i]:
    for b in graph[a]:
      for c in graph[b]:
        for person in graph[c]:
          graph[person] = []
          for name in graph[person]:
            name_list.remove(name)
          print(f"graph[{person}] = {graph[person]}")  
```

Now search for three people on each level and capture the time for each search.


```python
print(f"First level search names: {first_level_search}")
print(f"Second level search names: {second_level_search}")
print(f"Third level search names: {third_level_search}")
print(f"Fourth level search names: {fourth_level_search}")
print(f"Fifth level search names: {fifth_level_search}")
```

```python
time_list = []

level_list = [first_level_search, second_level_search, third_level_search, fourth_level_search, fifth_level_search]

for i in level_list:
    time_list.append(search_time(i, 0))
    time_list.append(search_time(i, 1))
    time_list.append(search_time(i, 2))
```

A breadth-first search algorithm looks for the shortest path or minimum number of edges between two nodes. BFS answers the questions of if there is a path from node A to node B and what the shortest path is from node A to node B. 

A five-level connections graph is created where each person in each level is connected to five other people. This graph is expressed through a hash table which establishes the relationships between a node and its neighbors. In the hash table or dictionary, each person is mapped to an array of its five neighbors in a time complexity of O(1). 

The breadth-first search algorithm searches through the first level before searching through the second level and the next level after that and on. The algorithm uses a First In, First Out data structure of queues to search through levels so that the first level is searched before the second. As each person is dequeued, the algorithm checks if the person is the target person. If not, that person's neighbors are added to the queue with a time complexity of O(1). That person's name is also added to a list so that the algorithm knows that the person was checked already. The search then loops to pop the next person in the queue to check if that person is the target person. 

As mentioned above, the time complexity of following each edge is O(1) and adding each node's neighbors to the queue is O(1). So the time complexity to follow the edges in a search is O(number of edges) and to adding to the queue is O(number of nodes). Therefore, the time complexity of the algorithm is O(V+E) where V is the number of vertices or nodes and E is the number of edges. This can be seen from the graph where the search for names in the third level took longer than that of the first or second level searches. This is a result of the algorithm running through more nodes an edges to get to the person in the third level than that of the first or second. Similarly, the search in the fourth level took longer than than the search in the third level and the fifth level search took longer than the fourth level search. To get to the deeper levels, the algorithm has to run through more nodes and edges, resulting in a larger O(V+E) than running through only the first or second levels. This trend can also be observed from the table where it breaks down the times for each person that was searched in each level. The time for each of three searches in the fifth level are greater than the times for any of the other searches in the other four levels. This algorihm will scale linearly with the increase in number of vertices and edges the algorithm runs through. 

This algorithm is useful because it can find if there is a path as well as search for the shortest path by looking at its neighbors first before going deeper. This is useful for data engineers to efficiently find the searched element using a scalable algorithm where its execution time will scale alongside the increase in load.
